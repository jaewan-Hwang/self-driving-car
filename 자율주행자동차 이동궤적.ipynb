```python
import numpy as np
import pandas as pd
import folium


cityName = str(input('차량 ID= '))
cityTour= df0[df0['차량 ID'].astype(str).str.contains(cityName)]


#위치중심

tourInfo= folium.Map(location=[36.454090,127.284024], zoom_start=13)

#지도에 표시
for i in range(len(cityTour)):
  location = cityTour.iloc[i,2], cityTour.iloc[i,3]
  marker= folium.CircleMarker(location,
                                tooltip='차량번호:'+'['+ str(cityTour.iloc[i,1])+']'+'프레임인덱스:'+ str(cityTour.iloc[i,0]),
                                radius=5, color='red', fill=True, fill_color='red')
  marker.add_to(tourInfo)

tourInfo
```
